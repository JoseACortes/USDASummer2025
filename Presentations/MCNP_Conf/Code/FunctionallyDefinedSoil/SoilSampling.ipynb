```python
# pick uniformly random points from a 1x1x1 cube and plot them in 3D
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
```

```python
# generate sets of random points, 10, 100, and 1000 points
num_points = [10, 100, 1000]
fig = plt.figure(figsize=(12, 8))
for i, n in enumerate(num_points):
    ax = fig.add_subplot(1, 3, i + 1, projection='3d')
    # generate random points
    x = np.random.rand(n)
    y = np.random.rand(n)
    z = np.random.rand(n)
    
    # plot the points
    ax.scatter(x, y, z, c='b', marker='o')
    # draw a cube around the points
    r = [0, 1]
    X, Y = np.meshgrid(r, r)
    # draw cube faces
    ax.plot_surface(X, Y, np.zeros_like(X), alpha=0.1, color='gray', edgecolor='black')      # bottom face z=0
    ax.plot_surface(X, Y, np.ones_like(X), alpha=0.1, color='gray', edgecolor='black')       # top face z=1
    ax.plot_surface(X, np.zeros_like(X), Y, alpha=0.1, color='gray', edgecolor='black')      # y=0 face
    ax.plot_surface(X, np.ones_like(X), Y, alpha=0.1, color='gray', edgecolor='black')       # y=1 face
    ax.plot_surface(np.zeros_like(X), X, Y, alpha=0.1, color='gray', edgecolor='black')      # x=0 face
    ax.plot_surface(np.ones_like(X), X, Y, alpha=0.1, color='gray', edgecolor='black')       # x=1 face
    
    # set labels
    # ax.set_xlabel('X axis')
    ax.set_xlim(-.5, 1.5)
    ax.set_xticks([])
    # ax.set_ylabel('Y axis')
    ax.set_ylim(-.5, 1.5)
    ax.set_yticks([])
    # ax.set_zlabel('Z axis')
    ax.set_zlim(-.5, 1.5)
    ax.set_zticks([])
    ax.set_title(f'Random Points: {n} points')
    ax.view_init(elev=0+45, azim=-90+45, )
    ax.set_aspect('equal')
plt.tight_layout()
plt.savefig('../../Figures/FunctionallyDefinedSoil/IndividualCellSampling.png', dpi=300, bbox_inches='tight')
plt.show()
```
