```python
import pandas as pd
import numpy as np
from tabulate import tabulate
import os
```

```python
import matplotlib.pyplot as plt
```

```python
figure_folder = '../../Figures/DataGeneration/'

os.makedirs(figure_folder, exist_ok=True)
```

```python
df = pd.read_pickle("DetectorReadings.pkl")
bins = df['bins'].values
df = df.drop(columns=['bins'])
df.index = bins
exp_df = pd.read_pickle('ExpirementData.pkl')
```

```python
exp_df
```

```python
feldspar_fn = exp_df[exp_df['function']=='Feldspar']['filename'].values[0]
feldspar_plot_info = {'filename': feldspar_fn, 'label': 'Feldspar', 'color': 'tab:blue'}
coconut_fn = exp_df[exp_df['function']=='Coconut']['filename'].values[0]
coconut_plot_info = {'filename': coconut_fn, 'label': 'Coconut', 'color': 'tab:green'}
sixty_percent_coconut_fn = exp_df[exp_df['function']=='C_1000_Feldspar_Fill']['filename'].values[0]
sixty_percent_coconut_plot_info = {'filename': sixty_percent_coconut_fn, 'label': r'10% Coconut 90% Feldspar', 'color': 'tab:orange'}
```

```python
plot_these = [feldspar_plot_info, coconut_plot_info, sixty_percent_coconut_plot_info]
for plot_info in plot_these:
    plt.plot(df.index, df[plot_info['filename']], label=plot_info['label'], color=plot_info['color'])

plt.axvline(x=4.44, color='k', linestyle='--', label='4.44 MeV (C-12)', alpha=0.5)
plt.xlabel('Energy (keV)')
plt.ylabel('Probability Density')
plt.yscale('log')
# plt.xscale('log')
plt.title('Detector Readings for Different Soil Types')
plt.legend()
plt.grid()


plt.savefig(figure_folder + 'DetectorReadings.png', dpi=100, bbox_inches='tight')
plt.show()
```
